## **Part 3: Reflection**

*In a brief paragraph, compare the two maps you created. Describe the visual and data model differences between the DEM datasets visualized in the topology map and the zonal statistics calculated and appended to the parcel dataset in the second map.*

**Upon looking at the two maps I made for this assignment, there are both visual and data model differences that visualize Oxford, Ohio. The visuals on the topography map are focused on the physical features of Oxford, as the transparent colors allow us to see the terrain of the city while also seeing the elevation level differences in the area. Meanwhile, the visuals utilize graduated colors to focus less on the physical characteristics of the area, but rather the levels of slope in parcel. The DEM dataset used in this topology map is representing elevation values across Oxford, Ohio, while the zonal statistics in the slope map calculated means of the raster values within defined zones of Oxford, Ohio.**
